# Receptor Cookbook

{class}`~pyrite.Receptor` in {mod}`pyrite` stores information about a protein. It allows you to load proteins, visualize them, and create and optimize docking workflows. On this page, we will get to know the Receptor object, and learn how to effectively use it.

Like {class}`~pyrite.Ligand`, {class}`~pyrite.Receptor` uses an rdkit {class}`~rdkit.Chem.rdchem.Mol` object to store chemical information. Unlike Ligand however, it is not an extension of Mol, but rather encapsulates a Mol object in its `rdkit` attribute. Therefore, we can still use rdkit methods, like {meth}`~rdkit.Chem.rdchem.Mol.GetNumHeavyAtoms`, on receptor:

```python
import sys
import os

bp = os.path.expanduser("~/Dev/bp")
sys.path.append(bp)
from pyrite import *
from rdkit import Chem
import numpy as np
```

```python
receptor = Receptor.from_pdb('input/4phl.pdb')
receptor.rdkit.GetNumHeavyAtoms()
```

```python
receptor.set_draw_options({'size': (800, 600)})
```
